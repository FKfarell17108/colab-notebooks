## 3. Exploratory Data Analysis (EDA)

### 3.1 EDA Univariate (Analisis Satu Variabel)

**Distribusi Skor Kemakmuran Rata-rata:**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['average_score'], bins=30, kde=True)
plt.title('Distribusi Average Score')
plt.xlabel('Skor Kemakmuran')
plt.ylabel('Jumlah Negara')
plt.grid(True)
plt.show()
```

Fungsi:
- Menampilkan **distribusi skor** kemakmuran rata-rata semua negara
- **histplot:** Membuat histogram
- **kde=True:** Menambahkan garis estimasi kepadatan

Cara Membaca:
- **Sebagian besar** negara memiliki **skor** antara **50-120**
- Distribusi tidak normal, **cenderung bimodal (memiliki dua puncak)**

**10 Negara Teratas dan Terbawah:**

```python
# 10 negara dengan skor tertinggi
top10 = df.nlargest(10, 'average_score')

# 10 negara dengan skor terendah
bottom10 = df.nsmallest(10, 'average_score')

plt.figure(figsize=(12, 5))
sns.barplot(x='average_score', y='country', data=top10, palette='Greens_r')
plt.title('10 Negara dengan Skor Kemakmuran Tertinggi')
plt.xlabel('Skor Kemakmuran')
plt.ylabel('Negara')
plt.show()

plt.figure(figsize=(12, 5))
sns.barplot(x='average_score', y='country', data=bottom10, palette='Reds_r')
plt.title('10 Negara dengan Skor Kemakmuran Terendah')
plt.xlabel('Skor Kemakmuran')
plt.ylabel('Negara')
plt.show()
```

Fungsi:
- **nlargest() dan nsmallest():** Mengambil n baris dengan nilai terbesar/terkecil
- **barplot:** Membuat diagram batang
- **palette:** Warna untuk diagram

Cara Membaca:
- Negara-negara **Eropa** mendominasi peringkat **teratas**
- Negara-negara **Afrika** dan konflik mendominasi peringkat **terbawah**

### 3.2 EDA Bivariate (Analisis Dua Variabel)

**Korelasi Antar Aspek Kemakmuran:**

```python
# Memilih hanya kolom numerik
numeric_cols = df.select_dtypes(include=['float64']).columns
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriks Korelasi Antar Aspek Kemakmuran')
plt.show()
```

Fungsi:
- **corr():** Menghitung matriks korelasi
- **heatmap:** Visualisasi matriks korelasi dengan warna
- **annot=True:** Menampilkan nilai korelasi

Cara Membaca:
- **Nilai mendekati 1:** Korelasi positif kuat
- **Nilai mendekati -1:** Korelasi negatif kuat
- **Nilai sekitar 0:** Tidak ada korelasi
- **Warna merah:** Korelasi positif
- **Warna biru:** Korelasi negatif

### 3.3 Analisis Indonesia

```python
indonesia = df[df['country'] == 'Indonesia'].iloc[0]

print("Peringkat Indonesia:", df[df['average_score'] <= indonesia['average_score']].shape[0], "dari", df.shape[0])
print("Skor Rata-rata Indonesia:", indonesia['average_score'])
print("Region:", indonesia['region'])
print("Political Regime:", indonesia['political_regime'])

# Visualisasi aspek kemakmuran Indonesia
indonesia_aspects = indonesia[numeric_cols].sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=indonesia_aspects.values, y=indonesia_aspects.index, palette='viridis')
plt.title('Aspek Kemakmuran Indonesia')
plt.xlabel('Skor')
plt.ylabel('Aspek')
plt.show()
```

Fungsi:
- Mengambil **data khusus Indonesia**
- Menghitung **peringkat Indonesia** berdasarkan skor rata-rata
- Membuat visualisasi aspek **kemakmuran Indonesia**

Output:
- Menampilkan **peringkat Indonesia** di antara 159 negara
- Menampilkan skor rata-rata dan aspek-aspek penting
- Diagram batang menunjukkan aspek terkuat dan terlemah Indonesia
